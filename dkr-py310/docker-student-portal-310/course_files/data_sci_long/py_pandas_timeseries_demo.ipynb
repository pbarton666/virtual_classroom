# Here is some pandas code in a iPython (Jupyter) notebook

```python
#py_pandas_timeseries_demo.ipynb
import numpy as np              
import pandas as  pd            
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#... and a few marker codes cf. matplotlib.org/api/colors_api.html
red='r'; blue='b'; green='g'; greenish='chartreuse'; magenta='m'
black='b';circle='o'; x='x';
```

This is what we'll actually plot.  We'll create new Series objects.

```python
#get some stuff to plot - permutations of a sine wave here...
#   ... these use list comprehensions and scalar operations
s1=pd.Series([np.sin(x/10) for x in range(0, 300)])
s2=s1.copy()
s2.index=s2.index-50
s3=s1*2
s1=pd.Series([np.cos(x/10) for x in range(0, 300)])
```

These commands plot the Series using matplotlib directly.

```python
#create a new figure (high-level container), add a plot to it
plt.figure()
plt.plot(s1)
```

We can jazz it up a bit with marrkers and colors ...

```python
plt.figure()
plt.plot(s1, red+circle, s1, black,
         s2, green+x, s2, blue,
         s3, greenish)
```

```python
#the series and it's stats
my_ser=pd.Series([3,4,5,2,7,5,3,2,8,5,4,3,2,5,6,7,1,1,1])
my_ser.sort_values(inplace=True)
mean=my_ser.mean()
std=my_ser.std()
N = my_ser.count()

#the corresponding normal distribution
#norm=pd.Series(np.random.normal(mean, std, N))

#creation of a histogram returns information about it.  "normed"
# means that SUM(col_height * col_width) = N
count, bins, ignored = plt.hist(my_ser, 20, normed=True)

#you can use optional settings for a line plot the first
#  argument is the probability density function
plt.plot(bins, 1/(std * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mean)**2 / (2 * std**2)) ,
          
         linewidth=1, 
         color='r')
plt
```

Optionally, we can stack several plots ('subplots') into a single container:

```python
rows=2; cols=2 #subplots fill row-wise

#we'll stack 'em in one at a time
plt.subplot(rows, cols, 1)
plt.plot(s1, red+circle, s1, black,
         s2, green+x, s2, blue)

plt.subplot(rows, cols, 2)
plt.plot(s2, green+x, s2, blue)

plt.subplot(rows, cols, 3)
plt.plot(s3, greenish, s1, black+circle)

plt.subplot(rows, cols, 4)
plt.scatter(s1, s2)
plt
```
