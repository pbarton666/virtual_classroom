```python
#py_matplot_compound.ipynb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fig, ax= plt.subplots(nrows=2, ncols=2)
fig
```

```python
ax
```

```python
ax[0,0].plot([1,2,3])
ax[0,1].plot([3,2,5])
ax[1,0].plot([1,4,3])
ax[1,1].plot([-1,-2,-3])
fig
```

```python
#drill down into individual subplots
sub00, sub01 = ax[0]   #first row
sub10, sub11 = ax[1]   #second row

sub11.scatter([2,.4,.5,0], [.2,.2,.3,1])
sub11.yaxis.set_ticks([-3, -2, -1, 0, 1, 2, 3, 4, 5])
sub11.xaxis.set_ticks([0,1,2])
sub11.set_xlabel("Yey ya!")
fig
```
